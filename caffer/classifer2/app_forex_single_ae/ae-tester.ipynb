```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.manifold import TSNE
import caffe
import math
import h5py
import random
caffe.set_mode_gpu()

def loadStockH5(fname):
    with h5py.File(fname, 'r') as f:   
        X = f['data'][:]
        y = f['label'][:] 
        y_ = f['realy'][:]
    return np.array(X).astype(np.float32), \
            np.array(y).astype(np.float32).reshape(len(y), 1, 1, 1), \
            np.array(y_).astype(np.float32)

%matplotlib inline
```

```python
datas, _, _ = loadStockH5("./caffe/data_1.h5")
print datas
gpu = 1

PRETRAINED_FILE = "m_%d_iter_17100.caffemodel" % gpu

dim_n = 1024
def forward(net, d, mask=dim_n):
    mask = int(mask)
    print mask,
    data = d.copy()
    #mean = data[0:mask].mean()
    #mean = data.mean()
    #print mean
    data = (data - data[mask-1]*0.999) * 1. * 255
    #data = (data - mean) * 1. * 255
    #data[mask:] = 0
    data[mask:] = data[mask]
    """
    for i in range(mask, data.shape[-1]):
        data.shape = (1, dim_n, 1, 1)
        net.blobs["indata"].data[...] = data
        net.forward()
        out = net.blobs["sigmoid_2"].data
        out.shape = (dim_n)
        data.shape = dim_n
        per = (out[i] - out[i-1]) / out[i-1]
        data[i:] = data[i-1] * per + data[i-1]
    """
    data.shape = (1, dim_n, 1, 1)
    net.blobs["indata"].data[...] = data
    net.forward()
    out = net.blobs["sigmoid_3"].data
    print "forward V2"
    #print net.blobs["drop2_1"].data
    #print out
    out.shape = (dim_n)
    return out.copy()
```

```python
def predict(deploy, model, split_per=0.8, i=-1):
    net = caffe.Net(deploy, model, caffe.TEST)
    if i < 0:
        i = random.randint(0, datas.shape[-1]-dim_n-1)
    d = datas[i:i+dim_n] / 10000
    split_n = int(split_per * dim_n)
    #d = np.zeros(dim_n)
    out1 = forward(net, d, split_n)
    del net
    print "finish", i
    fig = plt.figure(figsize=(12, 8))
    tmp = d[0:split_n]
    pridata = (d-tmp.min()) / (tmp.max()-tmp.min() + 0.00000001)
    plt.plot(range(dim_n), pridata, '-')
    plt.plot(range(dim_n), out1, '-')
    plt.plot([split_n,split_n],[0,1], color ='red', linewidth=2, linestyle="--")
    plt.legend(['0', '1', '2', '3', '4', '5'])
    plt.grid()
    plt.show()

```

```python
index = random.randint(0, datas.shape[-1]-dim_n-1)
print gpu, 
for j in range(80, 100):  
    predict("deploy_%d.prototxt" % gpu, PRETRAINED_FILE, split_per=j*1.0/100, i=index)
```
