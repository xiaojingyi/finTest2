```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.manifold import TSNE
import caffe
import math
import h5py
caffe.set_mode_gpu()

def loadStockH5(fname):
    with h5py.File(fname, 'r') as f:   
        X = f['data'][:]
        y = f['label'][:] 
        y_ = f['realy'][:]
    return np.array(X).astype(np.float32), \
            np.array(y).astype(np.float32).reshape(len(y), 1, 1, 1), \
            np.array(y_).astype(np.float32)

%matplotlib inline
```

```python
datas, _, _ = loadStockH5("./caffe/data_1.h5")
def makeOne(data):
    one = data.copy()
    tmp_1 = (one + abs(one))[:]
    tmp_2 = abs(one - abs(one))[:]
    tmp = np.concatenate((tmp_1, tmp_2)) / 2
    tmp.shape = (2048)
    return tmp

PRETRAINED_FILE = 'm_1_iter_17100.caffemodel' 

print datas
```

```python
def forward(net, data):
    net.blobs["indata"].data[...] = data
    net.forward()
    out = net.blobs["sigmoid_1"].data[0]
    out.shape = (2, 1024)
    out = out[0] - out[1]
    return out.copy()
net = caffe.Net("deploy.prototxt", PRETRAINED_FILE, caffe.TEST)
i = 50000
d = makeOne(datas[i:i+1024])
d = np.zeros(1024)
d = makeOne(d)
out1 = forward(net, d)
d = np.zeros(1024)
d[-1] = 0
d = makeOne(d)
out2 = forward(net, d)
tmp = out2.copy()
tmp[-1] = -0.9
d = makeOne(tmp)
out3 = forward(net, d)
del net
fig = plt.figure(figsize=(26, 18))
plt.plot(range(1024), out1, '-')
plt.plot(range(1024), out2, '-')
#plt.plot(range(1024), out3, '-')
d.shape = (2, 1024)
#plt.plot(range(1024), d[0]-d[1], '-')
print out2
print out3

plt.grid()
plt.show()
```

```python
net = caffe.Net("deploy.prototxt", PRETRAINED_FILE, caffe.TEST)
err_ct = same_ct = all_ct = 0
profit = potencial = 0
sum_all = 0
zero_sample = forward(net, np.zeros(2048))
for i in range(len(datas) - 1024 - 1):
    tmp = datas[i:i+1024].copy()
    last = tmp[-1]
    tmp[-1] = 0
    tmp = makeOne(tmp)
    predict = forward(net, tmp)
    final = predict[-1] - zero_sample[-1]
    sum_all += abs(final)
    if final * last < 0:
        err_ct += 1
        profit -= abs(last)
    elif final * last > 0:
        same_ct += 1
        profit += abs(last)
    potencial += abs(last)
    all_ct += 1
    if i % 10000 == 0:
        print same_ct, i, same_ct * 1.0 / all_ct, profit * 1.0 / potencial,
        print err_ct * 1.0 / all_ct#, sum_all/i
del net
```
