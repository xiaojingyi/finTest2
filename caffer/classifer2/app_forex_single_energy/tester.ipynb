```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.manifold import TSNE
import caffe
import math
import h5py
caffe.set_mode_gpu()

def loadStockH5(fname):
    with h5py.File(fname, 'r') as f:   
        X = f['data'][:]
        y = f['label'][:] 
        y_ = f['realy'][:]
    return np.array(X).astype(np.float32), \
            np.array(y).astype(np.float32).reshape(len(y), 1, 1, 1), \
            np.array(y_).astype(np.float32)

%matplotlib inline
```

```python
datas, _, _ = loadStockH5("./caffe/data_1.h5")
print datas
```

```python
MODEL_FILE = 'gpu1_stock.test.prototxt'
MODEL_TRAIN_FILE = 'gpu1_stock.test.prototxt'
PRETRAINED_FILE = 'm_1_iter_17100.caffemodel' 
net = caffe.Net(MODEL_FILE, PRETRAINED_FILE, caffe.TEST)
#net = caffe.Net(MODEL_TRAIN_FILE, PRETRAINED_FILE, caffe.TRAIN)
net.forward()
net.forward()
net.forward()
res1 = net.blobs["out_1"].data
data = net.blobs["data"].data
res1.shape = (256, 2048)
feat1 = net.blobs["sigmoid_1"].data
#print net.blobs["drop2_2"].data.shape
#print net.blobs["drop2_2"].data.sum()
#print net.blobs["drop2_3"].data.sum()

#print net.blobs["loss_debug"].data
#print net.blobs["concat_3"].data
#print net.blobs["data"].data
print (net.blobs["concat_3"].data == net.blobs["data"].data).sum() / 2048.
del net
```

```python
net = caffe.Net("deploy.prototxt", PRETRAINED_FILE, caffe.TRAIN)
err_ct = same_ct = all_ct = 0
profit = potencial = 0
sum_all = 0
for i in range(len(datas) - 1024 - 1):
    one = datas[i:i+1024].copy()
    last = one[-1]
    one[-1] = 0.
    tmp_1 = (one + abs(one))[:]
    tmp_2 = abs(one - abs(one))[:]
    tmp = np.concatenate((tmp_1, tmp_2)) / 2
    tmp.shape = (2048)
        
    net.blobs["data"].data[...] = tmp
    net.forward()
    out = net.blobs["sigmoid_1"].data[0]
    out.shape = (2, 1024)
    predict = out[1] - out[0]
    
    sum_all += abs(predict[-1])
    if abs(predict[-1]) < 0.0003:
        if predict[-1] * last < 0:
            err_ct += 1
            profit -= abs(last)
        elif predict[-1] * last > 0:
            same_ct += 1
            profit += abs(last)
    potencial += abs(last)
    all_ct += 1
    if i % 10000 == 0:
        print same_ct, i, same_ct * 1.0 / all_ct, profit * 1.0 / potencial,
        print err_ct * 1.0 / all_ct#, sum_all/i
del net
```
